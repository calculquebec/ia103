# Partie I: Qu'est-ce qu'une grappe de calcul

Pour l’instant nous avons été capables de tout faire avec l’équivalent en resources d’un ordi “normal”, mais et si…

- J’ai un modèle très massif a entrainer?

- J’ai des jeux de données massifs à traiter?

- Je veux faire de l’inference avec un très gros LLM?

- Je veux étudier plusieurs LLMs à la fois?

- Je veux entrainer un modèle de fondation?

## Le Calcul Informatique de Pointe

Avant de répondre à la question "Qu'est-ce qu'une grappe de calcul", définissons d'abord la classe de problèmes qui découle des questions que nous venons de poser. Ces scenarios sont "problematiques" sur un ordinateur normal seulement si l'ordinateur en question ne possede assez de resources comme: des coeurs de calcul, de la memoire vive ou du stockage, ou bien, si le temps que cela prendrait pour faire ces calculs est trop élevé

Nous allons dorénavant parler de ce genre de problème en utilisant le nom **calcul informatique de pointe**.

![calcul informtique de pointe](./images/cip.png)

Une **grappe de calcul** est un type d'équipement informatique utilisé dans pour traiter des problèmes de **calcul informatique de pointe**.

## Les Grappes de Calcul

Voici une définition assez directe de ce qu'est une grappe de calcul:

![definition grappe de calcul](./images/grappe_definition.png)

### Noeud de calcul

Ici, par **noeud de calcul** nous voulons dire un ordinateur avec un grand nombre de coeurs de calcul, ainsi qu'une grande quantité de mémoire vive et d'espace de stockage:

![un noeud de calcul](./images/noeud.png)

### Rack de calcul

Dans une **grappe de calcul**, plusieurs noeuds de calcul sont réliés par un réseau de haute performance. D'abord à l'intérieur d'un **rack**: un cabinet contenant des "tiroirs", où chaque tiroir est un noeud de calcul:

![un rack dans une grappe de calcul](./images/rack.png)

### Salle de serveurs

Ensuite, plusieurs racks sont à leur tour réliés par réseau de haute performance dans une **salle de serveurs**. Voilà notre **grappe de calcul**: ce sont des noeuds, réliés par réseau en plusieurs racks, eux-mêmes réliés par réseau dans une salle de serveurs.

![salle de serveurs](./images/salle.png)

# Comment utiliser nos grappes de calcul

La suite de cet atelier montrera comment utiliser les grappes de calcul de Calcul Québec, et celles de l'Alliance de Recherche Numérique du Canada (l'Alliance) en général. La premiére étape c'est de se connecter à une grappe de calcul.

## Connexion à nos grappes

Pour se connecter à une de nos grappes de calcul, la première étape c'est de créer un compte. Vous trouverez les instructions pour faire une demande de compte [sur ce lien](https://docs.alliancecan.ca/wiki/Apply_for_a_CCDB_account/fr). Pour la suite, nous allons faire comme si vous aviez déjà un compte actif avec l'Alliance.

À partir de votre ordinateur, vous ouvrirez une session *ssh* sur une de nos grappes. Par example, l'utilisateur <code>bob42</code>, qui désire se connecter sur la grappe <code>rorqual</code>, entrerait la commande suivant sur son ordinateur:

>$ ssh bob42@rorqual.calculquebec.ca

Si la commande fonctionne, et que <code>bob42</code> réussi à se conncter à la grappe <code>rorqual</code>, <code>bob42</code> se retrouverait sur un **noeud de login**, ou **noud de connexion**. Ce noeud n'est pas lui-même un noeud de calcul, même si des noeuds de calcul peuvent être plus costauds qu'un ordinateur typique: il est une porte d'entrée vers la grappe de calcul.

En d'autres mots, se connecter sur un noeud de login ne signifie pas que vous êtes immédiatement en mésure d'éxecuter un script python avec la commande:

> $ python train.py

et d'avoir accès automagiquement à plusieurs noeuds de calcul avec des centaines de GPU et des Tera-octets de mémoire. Il y aura d'autres étapes à suivre pour **soumettre des tâches de calcul** à la grappe, et cette partie du travail se fait dans la porte d'entrée de la grappe: le **noeud de login**.

![connexion a la grappe](./images/connexion.png)

En plus d'être le lieu à partir d'où vous allez **soumettre des tâches de calcul** à la grappe, le noeud de login rempli aussi d'autres rôles. Le noeud de login est l'endroit idéal pour:

- Configurer votre environnement de calcul
- Installer ou compiler des logiciels dont vous aurez besoin
- Faire des tests rapides et pas trop gros pour vous assurer que vous avez bien configuré votre environnement
- Gérer le stockage de vos résultats de calcul et des données dont vous avez besoin

Passons maintenant à comment faire pour **soumettre des tâches de calcul**.

## Tâches de calcul sur les grappes

Il y a deux types principaux de tâche de calcul:

1. Les tâches interactives
2. Les tâches "batch"

### Tâches interactives

Une tâche interactive, comme le nom le suggère, est une tâche où l'utilisateur demande une certaine quantité de ressources sur un noeud de calcul et souhaite travailler de manière interactive. C'est à dire, l'utilisateur souhaite faire des choses tel que: exécuter des commandes manuellement sur une ligne de commande, éditer des scripts et les exécuter rapidement après chaque changement ou faire de la visualization de données en exécutant manuellement des lignes de code.

Sur nos grappes de calcul, il y a deux façons de faire une requête de **tâche interactive**. La première est par voie de la commande <code>salloc</code>. Admettons que l'utilisateur <code>bob42</code> souhaite exécuter une suite de scripts python pour vérifier si son code est bel et bien en mésure d'utiliser un GPU avant de soummetre une longue tâche de calcul de plusieurs heures. Il a besoin de 64G de mémoire vive, ainsi que de 8 coeurs CPU et 1 GPU pour faire ses tests. Il pense avoir besoin de moins d'une heure pour tout tester. Il peut alors, à partir d'un noeud de login, utiliser la commande <code>salloc</code> comme ceci:

> $ salloc --account=def-bobsprof --ntasks=1 --mem=64G --cpus-per-task=8 --gres:gpu:1 --time=00:01:00

Après un certain temps d'attente, il obtiendra un terminal sur un noeud de calcul où il aura accès à exactement la quantité de ressources qu'il a demandé, pendant la durée de temps qu'il a demandé.

Vous trouverez plus de détails sur cette commande, y compris quels paramètres de ressources sont disponibles, [sur ce lien](https://docs.alliancecan.ca/wiki/Running_jobs/fr#Tâches_interactives)

La deuxième façon est celle que nous utilisons à présent. Cet environnement Jupyter où nous nous trovons est en fait une tâche interactive sur une grappe de calcul à la quelle nous nous connectons par notre navigateur web.

Si vous avez un compte avec l'Alliance, vous pouvez par exemple lancer une session Jupyter sur la grappe rorqual sur l'addresse: [https://jupyterhub.beluga.computecanada.ca/](https://jupyterhub.beluga.computecanada.ca/). En dessous du capot, il se passe ceci:

!["jupyerlab session on rorqual"](./jupyterlab.png)

Vous trouverez plus d'informations, comme les grappes où il est possible de se connecter à Jupyter par votre navigateur web, [sur ce lien](https://docs.alliancecan.ca/wiki/JupyterLab/fr)

### Tâches Batch

Contrairement à une tâche interactive, une tâche "batch" est une tâche où nous envoyons des instructions à la grappe de calcul et nous n'avons pas besoin de toucher à quoi que ce soit une fois que c'est envoyé. Nous voulons seulement exécuter une ou plusieurs commandes dans un (ou plusieurs) noeud de calcul avec suffisament de ressources, et nous vérifierons les résultats plus tard une fois que les calculs seront finis.

Pour **soumettre une tâche batch**, vous pouvez, à partir d'un noeud de login, utiliser la commande <code>sbatch</code> pour envoyer un **script de tâche** à la grappe:

> $ sbatch ma_tache.sh

Un **script de tâche** est un script *shell* qui informe la grappe de la quantité de ressources dont vous voulez faire l'utilisation, ainsi que quelles sont les commandes que vous souhaitez exécuter. Un script de tâche a deux sections: l'entête, où nous écrivons la quantité de ressources desirée; et le corps, où nous écrivons les commandes qui devront être exécutées dans l'ordre dans le quel nous voulons les exécuter:

![example de script de tâche](./images/job_script.png)

Vous trouverez plus de détails sur ce sujet [sur ce lien](https://docs.alliancecan.ca/wiki/Running_jobs)

Quel que soit le type de tâche, la mécanique derrière est la même: un utilisateur envoie une requête de tâche à un **ordonnanceur**, un logiciel qui décide où placer une tâche sur la grappe selon la demande en ressources et la **priorité** de l'utilisateur, et la tâche démarre sur un noeud de calcul sur la grappe:

![grappe de calcul](./images/cluster.png)
