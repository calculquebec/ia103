## Téléchargement des ressources de Huggingface

### Transformers
Transformers est un package Python qui fournit des API et des outils permettant de télécharger et d'entraîner facilement des modèles de pointe, pré-entraînés sur diverses tâches dans de multiples domaines.

### Installation des transformers

Nous vous recommandons de l'installer à l'aide du package Python fourni, comme suit :

1. Chargez le module Python avec la commande `module load python`.
2. Créez et démarrez un environnement virtuel.
3. Installez Transformers dans l'environnement virtuel avec la commande `pip install`.


<code>pip install --no-index transformers</code>

### Téléchargement de modèles pré-entraînés

Plusieurs méthodes permettent de télécharger un modèle pré-entraîné du hub de modèles HuggingFace :

#### Utilisation de git lfs

Les modèles pré-entraînés sont généralement composés de fichiers binaires assez volumineux. HuggingFace met ces fichiers à disposition via Git Large File Storage. Pour télécharger un modèle, chargez le module git-lfs et clonez le dépôt du modèle choisi du hub :

<code>module load git-lfs/3.4.0 </code><br>
<code>git clone --depth 1 --jobs 1 https://huggingface.co/bert-base-uncased </code>

Maintenant que vous disposez d'une copie du modèle pré-entraîné enregistrée localement dans le système de fichiers de la grappe de calcul, vous pouvez la charger avec un script Python dans une tâche comportant l'option <code>local_files_only</code> afin d'éviter les tentatives de téléchargement du Web :

```bash
transformers import AutoModel, AutoTokenizer
model = AutoModel.from_pretrained("/path/to/where/you/cloned/the/model", local_files_only=True)
tokenizer = AutoTokenizer.from_pretrained("/path/to/where/you/cloned/the/model", local_files_only=True)

#### Utilisation de l'interface de ligne de commande Hugging Face

Le paquet huggingface_hub contient une interface de ligne de commande (CLI) permettant de télécharger des modèles. Par exemple, pour télécharger le modèle Zephyr-7b-beta, installez d'abord huggingface_hub dans un environnement virtuel, puis exécutez la commande suivante sur un nœud de connexion :

<code>HF_HUB_DISABLE_XET=1 hf download --max-workers=1 HuggingFaceH4/zephyr-7b-beta</code>

#### Utiliser Python
Il est également possible de télécharger des modèles pré-entraînés en utilisant Python plutôt que Git. La commande suivante doit être exécutée sur un nœud de connexion, car une connexion Internet est requise pour télécharger les fichiers du modèle :

```bash
from transformers import AutoModel, AutoTokenizer
model = AutoModel.from_pretrained("bert-base-uncased")
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

Cela stockera les fichiers du modèle pré-entraîné dans un répertoire cache, par défaut $HOME/.cache/huggingface/hub. Vous pouvez modifier le répertoire cache en procédant comme suit :
* Définissez la variable d'environnement TRANSFORMERS_CACHE avant d'importer quoi que ce soit du package transformers dans votre script Python.
* passer `cache_dir=/path` à `'AutoTokenizer.from_pretrained`.

Une fois le modèle téléchargé, vous pouvez toujours le charger du disque local (que ce soit le cache par défaut ou un autre emplacement) :
```bash
from transformers import AutoModel, AutoTokenizer
model = AutoModel.from_pretrained("/path/to/where/model/is/saved", local_files_only=True)
tokenizer = AutoTokenizer.from_pretrained("/path/to/where/model/is/saved", local_files_only=True)

#### Utiliser le pipeline Python
Une autre méthode courante pour charger un modèle pré-entraîné consiste à utiliser un pipeline. Sur un nœud de connexion, il suffit de transmettre un nom de modèle ou un type de tâche comme argument à la commande `pipeline`. Le modèle sera alors téléchargé et stocké à l'emplacement de cache par défaut.

```bash
from transformers import pipeline
pipe = pipeline("text-classification",model="bert-base-uncased")

Une fois encore, une fois le modèle enregistré sur le disque, dans un environnement sans connexion Internet, comme au sein d'une tâche, vous devez spécifier son emplacement ainsi que celui de son tokenizer lors de l'appel du pipeline :
```bash
from transformers import pipeline, AutoModel, AutoTokenizer
 model = AutoModel.from_pretrained("/path/to/where/model/is/saved", local_files_only=True)
 tokenizer = AutoTokenizer.from_pretrained("/path/to/where/model/is/saved", local_files_only=True)
 pipe = pipeline(task = "text-classification", model = model, tokenizer = tokenizer)

### Datasets
Datasets est un package Python permettant d'accéder et de partager facilement des ensembles de données pour les tâches audio, de vision par ordinateur et de traitement automatique des tâches du Natural Language Processing (NLP).

### Installation des Datasets
Nous vous recommandons de l'installer à l'aide du package Python fourni, comme suit :

1. Chargez le module Python : `module load python`

2. Créez et démarrez un environnement virtuel.

3. Chargez le module Arrow. Le package pyarrow (une dépendance de Datasets) sera ainsi disponible dans votre environnement virtuel.

4. Installez Datasets dans l'environnement virtuel avec `pip install`.

```bash
module load gcc/9.3.0 arrow/11.0.0
pip install --no-index datasets

### Téléchargement des Datasets
La méthode exacte pour télécharger et utiliser des Datasets du hub Hugging Face dépend de plusieurs facteurs, tels que le format et le type de tâche pour laquelle les données seront utilisées. Quelle que soit la méthode choisie, tout téléchargement doit être effectué sur un nœud connecté. Consultez la documentation officielle du package pour plus de détails sur le téléchargement des différents types de jeux de données.

Une fois le jeu de données téléchargé, il est stocké localement dans un répertoire cache, par défaut `$HOME/.cache/huggingface/datasets`. Vous pouvez modifier l'emplacement du cache par défaut en définissant la variable d'environnement `HF_DATASETS_CACHE` avant d'importer des éléments du package `Datasets` dans votre script Python.

Pour charger un jeu de données dans une tâche sans connexion Internet, définissez la variable d'environnement `HF_DATASETS_OFFLINE` sur 1 et spécifiez l'emplacement du répertoire cache où le jeu de données est stocké lors de l'appel à `load_dataset()`.

```python 
import os
os.environ['HF_DATASETS_OFFLINE'] = '1'
from datasets import load_dataset
dataset = load_dataset("/path/to/loading_script/of/the/dataset")

### Accelerate
Accelerate est un package qui permet d'exécuter n'importe quel code PyTorch sur n'importe quelle configuration distribuée en ajoutant seulement quatre lignes de code. Cela simplifie, optimise et rend adaptable l'entraînement et l'inférence à grande échelle. 

Vous pouvez ainsi utiliser le même code d'entraînement, avec des modifications minimes, sur un seul processeur, plusieurs GPU et TPU, sans avoir à maîtriser des API d'entraînement distribué complexes comme torch.distributed.


### Installation d'Accelerate
Nous vous recommandons de l'installer à l'aide de notre roue d'installation Python fournie, comme suit :

1. Chargez le module Python : `module load python`

2. Créez et démarrez un environnement virtuel.

3. Installez Accelerate dans l'environnement virtuel avec `pip install`.

```bash
pip install --no-index accelerate

### Accélérer le code PyTorch avec Accelerate

**accelerate.sh:**
```bash
#!/bin/bash
#SBATCH --nodes 2
#SBATCH --ntasks-per-node=1
#SBATCH --gpus-per-task=2
#SBATCH --cpus-per-task=4 
#SBATCH --mem=16000M       
#SBATCH --time=0-00:10
#SBATCH --output=%N-%j.out

## Create a virtualenv and install accelerate + its dependencies on all nodes ##
srun -N $SLURM_NNODES -n $SLURM_NNODES config_env.sh

export HEAD_NODE=$(hostname) # store head node's address
export HEAD_NODE_PORT=34567 # choose a port on the main node to start accelerate's main process

srun launch_training_accelerate.sh

***config_env.sh:***
```bash
#!/bin/bash

module load python

virtualenv --no-download $SLURM_TMPDIR/ENV

source $SLURM_TMPDIR/ENV/bin/activate

pip install --upgrade pip --no-index

pip install --no-index torchvision accelerate

echo "Done installing virtualenv!"

**launch_training_accelerate.sh:**
```bash
#!/bin/bash

source $SLURM_TMPDIR/ENV/bin/activate
export NCCL_ASYNC_ERROR_HANDLING=1

echo "Node $SLURM_NODEID says: main node at $HEAD_NODE"
echo "Node $SLURM_NODEID says: Launching python script with accelerate..."

accelerate launch \
--multi_gpu \
--gpu_ids="all" \
--num_machines=$SLURM_NNODES \
--machine_rank=$SLURM_NODEID \
--num_processes=4 \ # This is the total number of GPUs across all nodes
--main_process_ip="$HEAD_NODE" \
--main_process_port=$HEAD_NODE_PORT \
pytorch-accelerate.py --batch_size 256 --num_workers=2

Accelerate extrait précisément le code répétitif relatif aux GPU multiples, aux TPU et à la précision fractionnaire (fp16), laissant le reste de votre code inchangé.

Comme vous pouvez le constater, l'ajout de cinq lignes à n'importe quel script d'entraînement PyTorch standard permet désormais de l'exécuter sur n'importe quelle configuration de nœud (monoprocesseur, GPU unique, multi-GPU et TPU), avec ou sans précision mixte (fp8, fp16, bf16).

Ce même code peut ensuite être exécuté sans modification sur votre machine locale pour le débogage ou dans votre environnement d'entraînement.
