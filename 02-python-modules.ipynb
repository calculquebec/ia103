## Comment charger des modules Python et créer un environnement virtuel

Ici on parle de l'interpréteur Python, ainsi que la manière d'installer et d'utiliser les paquets.

### Version Python par défaut

Lorsque vous vous connectez à nos grappes de calcul, une version Python par défaut est disponible, mais vous pouvez installer une autre version qui correspond à vos besoins.

### Chargement d'un module Python

Pour découvrir les versions de Python disponibles:

`module avail python`

Pour charger le module de votre choix :

<code>module load python/X.Y</code>

### How to create and use a virtual environment

Pour créer un environnement virtuel, assurez-vous d'avoir sélectionné une version de Python avec l'option `module load python/X.Y.Z` comme indiqué ci-dessus. Saisissez ensuite la commande suivante, en remplaçant `ENV` par le nom du répertoire de votre nouvel environnement :

<code>virtualenv --no-download ENV</code>

Une fois l'environnement virtuel créé, il faut l'activer :

<code>source ENV/bin/activate</code>


Vous devriez également mettre à jour pip dans l'environnement :

<code>pip install --no-index --upgrade pip</code>

Pour quitter l'environnement virtuel, il suffit de saisir la commande disable :

<code>deactivate</code>

*--no-download* (or --never-download) Cette option empêche l'outil de se connecter à Internet pour récupérer les dernières versions de pip, setuptools ou wheel lors de la création d'un nouvel environnement virtuel. Il utilise plutôt les paquets inclus localement dans l'installation.

*--no-index* Cette option empêche pip de se connecter à l'index des paquets Python (PyPI) par défaut pour télécharger ou rechercher des paquets. Elle force l'installateur à ne consulter que les fichiers locaux ou les dépôts personnalisés explicitement définis.

### Comment installer d'autres paquets

La première étape consiste à charger un module Python et à activer un environnement virtuel (ENV). Ensuite, nous installons par exemple le package NumPy :

```bash
module load python/X.Y
source ENV/bin/activate</code>
pip install numpy --no-index

* La commande pip permet d'installer des paquets provenant de diverses sources, notamment PyPI et les distributions précompilées appelées roues Python. Nous fournissons des roues Python pour plusieurs paquets. Dans l'exemple ci-dessus, l'option `--no-index` indique à pip de ne pas installer de paquets depuis PyPI, mais uniquement ceux disponibles localement, c'est-à-dire nos roues. <br>
* Si vous omettez l'option `--no-index`, pip recherchera à la fois dans PyPI et dans les paquets locaux, et utilisera la dernière version disponible. Si une version plus récente est disponible sur PyPI, elle sera installée à la place de notre roue, ce qui peut entraîner des problèmes.

### Sujet avancé : Enregistrez votre environnement virtuel en tant que noyau Jupyter
Il est parfois bon idee de créer des noyaux personnalisés dans Jupyter. Cela permet de gérer les exigences de différents projets et environnements de langage sans créer de conflits au sein du système. L'ajout d'un noyau personnalisé permet à une seule interface Jupyter d'interagir avec plusieurs configurations isolées.

```bash
mkdir -p ~/.local/share/jupyter/kernels
pip install --no-index ipykernel
python -m ipykernel install --user --name Python3.12_TestKernel --display-name "Python 3.12 Test Kernel"

Ouvrez maintenant une fenêtre de lancement et ouvrez une icône Jupyter Notebook avec votre noyau personnalisé.
